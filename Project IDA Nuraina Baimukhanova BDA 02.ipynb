### 3.2. Data manipulation: cleaning and shaping

In this step, we do following manipulations:
+ Inconsistent column names 
+ Dropping unnecessary columns
+ Missing data
+ Different data types
+ Duplicated rows
+ Data Shaping


#### Inconsistent column names

```python
#check the column names and their data type
data.columns
```

```python
#change the case to upper
data.columns=data.columns.str.upper()
data.columns
```

```python
#rename columns 
data.rename(columns={'PDDISTRICT':'DISTRICT','DAYOFWEEK':'DAY OF WEEK'},inplace = True);
data.head()
```

#### Dropping unnecessary columns

The first and last columns, called incident number(INCIDNTNUM) and police department id(PDID) respectievly, will not be obliging and helpful in analysis. 

```python
data.drop(columns=['INCIDNTNUM', 'PDID'],inplace = True);
data
```

#### Missing data

```python
#checking for the missing data
#false means no missing data  
data.isnull().any()
```

```python
#columns with total row numbers, which are not empty 
data1 = pd.DataFrame()
data1["Columns"] = list(data.columns)
data1["Count"] = list(data.count())

data1
```

```python
#total number of missed values
data.isnull().sum().sum()
```

```python
#add default value for missed data
data=data.fillna('-')
#deleting rows with NaN values
#data = data.dropna()
```

```python
#number of missed value after getting rid/replacing 
data.isnull().sum().sum()
```

#### Duplicated data

```python
#checking for duplicated data
#shows total number of duplicated data
data.duplicated().sum()
```

```python
#retrieve all duplicated rows 
data.loc[data.duplicated(),:]
```

```python
#drop duplicates except the very first row 
data=data.drop_duplicates(keep='first')
```

```python
# the number of rows left after dropping duplicates
data.shape
```

#### Data types

```python
#checking the data types 
data.columns
```

```python
#assigning the appropriate data type
data["DATE"] = data["DATE"].apply(pd.to_datetime)
data["TIME"] = data["TIME"].apply(pd.to_datetime)
```
